```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['figure.figsize'] = (20, 10)
sns.set(rc={'figure.figsize':(20,10)})
```

## Training data

```python
df = pd.read_csv('train_v9rqX0R.csv')
df
```

```python
df.nunique()
```

### Preprocessing of data

```python
df['Item_Fat_Content'].unique()
```

```python
def process_Item_Fat(x):
    if(x == 'low fat'):
        x = x.replace('low fat', 'Low Fat')
    if(x == 'LF'):
        x = x.replace('LF', 'Low Fat')
    if(x == 'reg'):
        x = x.replace('reg', 'Regular')
    return x

df['Item_Fat_Content'] = df['Item_Fat_Content'].apply(process_Item_Fat)
```

```python
df['Item_Fat_Content'].unique()
```

### Filling null value in Item_Weight column

```python
df.isnull().sum()
```

```python
# fill the missing values
item_weight_mean = df.pivot_table(values = "Item_Weight", index = 'Item_Identifier')
item_weight_mean
```

```python
miss_value_bool = df['Item_Weight'].isnull()
miss_value_bool
```

```python
for i, item in enumerate(df['Item_Identifier']):
    if miss_value_bool[i]:
        if item in item_weight_mean:
            df['Item_Weight'][i] = item_weight_mean.loc[item]['Item_Weight']
        else:
            df['Item_Weight'][i] = np.mean(df['Item_Weight'])
```

```python
df.isnull().sum()
```

### Reduce zero value from Item_Visibility column

```python
sum(df['Item_Visibility']==0)
```

```python
# replace zeros with mean
df.loc[:, 'Item_Visibility'].replace([0], [df['Item_Visibility'].mean()], inplace=True)
```

```python
sum(df['Item_Visibility']==0)
```

### Creating new column from Item_Identifier column

```python
df['New_Item_Type'] = df['Item_Identifier'].apply(lambda x: x[:2])
df['New_Item_Type']
```

```python
df['New_Item_Type'] = df['New_Item_Type'].map({'FD':'Food', 'NC':'Non-Consumable', 'DR':'Drinks'})
df['New_Item_Type'].value_counts()
```

```python
# according to new_item_type column we update the value in Item_Fat_Content column
```

```python
df.loc[df['New_Item_Type']=='Non-Consumable', 'Item_Fat_Content'] = 'Non-Edible'
df['Item_Fat_Content'].value_counts()
```

### Handle Null value in Outlet_Size column

```python
# log transformation
df['Item_Outlet_Sales'] = np.log(1+df['Item_Outlet_Sales'])
```

```python
df1 = df.dropna(how = 'any')
```

```python
df1.head()
```

```python
df2 = df[df['Outlet_Type'] == 'Supermarket Type1']
df3 = df2[df2['Outlet_Location_Type'] == 'Tier 2']
```

```python
df3
```

```python
df3.isnull().sum()
```

```python
df.Outlet_Type.unique()
```

```python
df4 = df[df.Outlet_Type == 'Grocery Store']
df5 = df4[df4.Outlet_Location_Type == 'Tier 3']
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits = 10, test_size = 0.2, random_state = 0)
cross_val_score(LinearRegression(), x, y, cv = cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_GSCV(x, y):
    algos = {
        'linear_regression' : {
            'model' : LinearRegression(),
            'params' : {
                'normalize' : [True, False]
            }
        },
        'lasso' : {
            'model' : Lasso(),
            'params' : {
                'alpha' : [1, 2],
                'selection' : ['random', 'cyclic']
            }
        },
        'decision_tree' : {
            'model' : DecisionTreeRegressor(),
            'params' : {
                'criterion' : ['mse', 'friedman_mse'],
                'splitter' : ['best', 'random']
            }
        }
    }
    
    scores = []
    cv = ShuffleSplit(n_splits = 10, test_size = 0.2, random_state = 0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv = cv, return_train_score = False)
        gs.fit(x, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
        
    return pd.DataFrame(scores, columns = ['model', 'best_score', 'best_params'])
```

```python
find_best_model_using_GSCV(x, y)
```
